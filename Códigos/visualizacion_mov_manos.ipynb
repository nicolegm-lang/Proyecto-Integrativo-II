### **Visualización de señales**

Las señales IMU por prueba obtenidas del sensor WT9011DLC constan de señales de aceleración, velocidad angular, ángulo y cuarteniones en el eje *x, y, z* de cada uno.

El dataset originalmente también contiene columnas *Time, Device name, Chip time, Shift, SpeedX, Magnetic Field y Temperature* pero no son relevantes para el análisis por lo que, no se graficaron dichas columnas.


Para tener un primera observación de las señales, se graficaron las siguientes señales correspondientes a la prueba de un sujeto aleatorio.

***Aceleración***

```python
import matplotlib.pyplot as plt
import random

# Carpeta de la clase AB
ruta_clase_ab = os.path.join(path, 'AB')

# Lista de archivos CSV en la carpeta AB
archivos_ab = glob.glob(os.path.join(ruta_clase_ab, "*.csv"))

# Seleccionamos un archivo aleatorio
archivo_aleatorio = random.choice(archivos_ab)

# Cargamos el archivo CSV en un DataFrame
df_aleatorio = pd.read_csv(archivo_aleatorio, on_bad_lines='skip')

# Grafica de las señales de aceleración
plt.figure(figsize=(12, 6))

plt.plot(df_aleatorio['Acceleration X(g)'], label='Acceleration X(g)')
plt.plot(df_aleatorio['Acceleration Y(g)'], label='Acceleration Y(g)')
plt.plot(df_aleatorio['Acceleration Z(g)'], label='Acceleration Z(g)')

plt.title(f'Señales de Aceleración de {os.path.basename(archivo_aleatorio)}')
plt.xlabel('Muestras')
plt.ylabel('Aceleración (g)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Carpeta de la clase ID
ruta_clase_id = os.path.join(path, 'ID')

# Lista de archivos CSV en la carpeta ID
archivos_id = glob.glob(os.path.join(ruta_clase_id, "*.csv"))

# Seleccionamos  un archivo aleatorio
archivo_aleatorio_id = random.choice(archivos_id)

# Cargamos el archivo CSV en un DataFrame
df_aleatorio_id = pd.read_csv(archivo_aleatorio_id, on_bad_lines='skip')

# Grafica de las señales de aceleración
plt.figure(figsize=(12, 6))

plt.plot(df_aleatorio_id['Acceleration X(g)'], label='Acceleration X(g)')
plt.plot(df_aleatorio_id['Acceleration Y(g)'], label='Acceleration Y(g)')
plt.plot(df_aleatorio_id['Acceleration Z(g)'], label='Acceleration Z(g)')

plt.title(f'Señales de Aceleración de {os.path.basename(archivo_aleatorio_id)}')
plt.xlabel('Muestras')
plt.ylabel('Aceleración (g)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Carpeta de la clase SM
ruta_clase_sm = os.path.join(path, 'SM')

# Lista de archivos CSV en la carpeta SM
archivos_sm = glob.glob(os.path.join(ruta_clase_sm, "*.csv"))

# Seleccionamos un archivo aleatorio
archivo_aleatorio_sm = random.choice(archivos_sm)

# Cargamos el archivo CSV en un DataFrame
df_aleatorio_sm = pd.read_csv(archivo_aleatorio_sm, on_bad_lines='skip')

# Grafica de las señales de aceleración
plt.figure(figsize=(12, 6))

plt.plot(df_aleatorio_sm['Acceleration X(g)'], label='Acceleration X(g)')
plt.plot(df_aleatorio_sm['Acceleration Y(g)'], label='Acceleration Y(g)')
plt.plot(df_aleatorio_sm['Acceleration Z(g)'], label='Acceleration Z(g)')

plt.title(f'Señales de Aceleración de {os.path.basename(archivo_aleatorio_sm)}')
plt.xlabel('Muestras')
plt.ylabel('Aceleración (g)')
plt.legend()
plt.grid(True)
plt.show()
```

En las señales de AB la acelación en el eje X se mantiene alrededor de cero, mientras que en Y inicia con valores arriba de 1g y oscila aproximadamente entre 0.5 y 1.5g. En Z oscila cruzando por cero en distintas amplitudes pero aproximadamente entre -1 y arriba de 0.5.

Para la señal ID, en el eje X y Y la aceleración es nula, únicamente varía en eje Z cruzando por cero.

Como era de esperarse, en la señal SM en los tres ejes la aceleración es casi nula.

*Anotaciones:*

* Se destaca la factibilidad de extracción de características para la señal en Y y Z correspondiente a AB y de la señal en Z en ID.
* Si la aceleración en X y Y es nula, corresponde a señal ID.
* Si la aceleración en X,Y,Z es nula, corresponde a señal SM.
